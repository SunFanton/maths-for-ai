## Vectores en Python

<p style="font-size:1.1rem">En Python, un vector puede representarse de muchas maneras, la más común y simple es mediante una lista de números normal de Python. Por ejemplo:</p>

```python
v = [10.5, 5.2, 3.25, 7.0] #Esto es una lista normal de numeros de Python
v
```

<p style="font-size:1.1rem">Pero para la gran cantidad de cálculos matemáticos y científicos que se requieren en Ciencia de Datos e Inteligencia Artificial, lo más conveniente es utilizar las implementaciones de vectores y operaciones optimizadas que nos provee la librería <b>NumPy</b>, ya que Python suele ser más lento para esto. Por otro lado, NumPy corre sobre C y C++, lo que favorece a su rápidez de ejecución. Por ejemplo:</p>

```python
import numpy as np
video = np.array([10.5, 5.2, 3.25, 7.0]) # esta es la manera de generar arreglos en NumPy, de la dimension que sea
print(video) # imprime el valor del vector
video # imprime el vector en si
```

<p style="font-size:1.1rem">El tamaño de un vector (cantidad de componentes o elementos del arreglo) lo puedo conocer mediante el atributo <b>size</b>:</p>

```python
video.size
```

<p style="font-size:1.1rem">Para acceder a los diferentes elementos del vector, necesitamos utilizar índices. En Matemáticas, dichos índices comienzan en 1, mientras que en programación empiezan en 0:</p>

```python
video[2] # accedemos al tercer elemento del vector
```

## Graficar vectores

<p style="font-size:1.1rem">Si bien hemos visto que existe una diferencia en el concepto de lo que es un vector desde el punto de vista matemático y desde el punto de vista de la informática, aún así podemos utilizar Python para graficar vectores, tal cual los conocemos en matemáticas. Para esto, se suele utilizar la librería <b>matplotlib</b>.</p>

### Importamos la librería

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

### Introducción a matplotlib

<p style="font-size:1.1rem">Los siguientes ejemplos corresponden a algunas de las funciones más básicas de <b>matplotlib</b>. Dichos ejemplos hasn sido extraídos del material didáctico de <a href="https://www.w3schools.com/python/matplotlib_intro.asp">W3Schools</a></p>

#### Ejemplo 1

```python
# Forma basica de generar un grafico en matplotlib (esto es para conocer la libreria, no estamos representando un vector)

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints) # Por defecto, plot() dibuja una linea de un punto a otro de acuerdo a los parametros de coordenadas recibidos
plt.show() # La funcion show() es fundamental para efectivamente mostrar el grafico
```

#### Ejemplo 2

```python
# Dibujamos solo los puntos, no la linea que los une, gracias al tercer parametro

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

#### Ejemplo 3

```python
# Dibujamos las lineas que unen ya no dos puntos, sino por ejemplo cuatro

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

#### Ejemplo 4

```python
'''
En este ejemplo, por un lado solo estamos pasando las coordenadas del eje y. Esto es posible ya que matplotlib 
considera que las coordenadas del eje x para cada una de las del eje y correponderan a la secuencia 0, 1, 2, 3...
Por otra parte, con el parametro marker estamos indicando que los puntos se resaltaran con un circulo.
ms -> short name para markersize
mec -> short name para markeredgecolor
mfc -> short name para marker face color
'''

ypoints = np.array([3, 8, 1, 10]) # las coordenadas del eje x seran 0, 1, 2, 3

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'red', mfc = 'white') 
plt.show()
```

#### Ejemplo 5

```python
'''
Se pueden crear lineas de distintos estilos.
De alli el uso del parametro linestyle (abreviado es ls)
color -> abreviado es c
linewidth -> abreviado es lw
Tambien se puede dibujar mas de una linea en el mismo grafico
'''

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, linestyle = "dashed", c = "hotpink", lw = "5")
plt.plot(y2)

plt.show()
```

#### Ejemplo 6

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

#### Ejemplo 7

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5) # Para agregar una cuadricula de fondo con propiedades personalizadas

plt.show()
```

#### Ejemplo 8

```python
# La funcion scatter() dibuja por defecto los puntos que le pasemos como parametros en las coordenas

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show()
```

### Vectores 2D

#### Ejemplo 1

```python
u = np.array([2, 5])
v = np.array([3, 1])

x_coords, y_coords = zip(u, v)
plt.scatter(x_coords, y_coords, color=["r","b"])
plt.axis([0, 9, 0, 6]) # indican los valores minimo y maximo que se mostraran en ambos ejes
plt.grid()
plt.show()
```
